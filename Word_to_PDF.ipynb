```python
import os
from google.colab import files

# Install LibreOffice
!apt-get install -y libreoffice

def convert_word_to_pdf(input_file, output_file):
    """
    Convert a Word document to PDF using LibreOffice command-line tool.
    """
    command = f"libreoffice --headless --convert-to pdf '{input_file}' --outdir '{os.path.dirname(output_file)}'"
    os.system(command)

def upload_and_convert():
    print("Upload your Word document:")
    uploaded = files.upload()

    for file_name in uploaded.keys():
        if file_name.endswith('.docx'):
            output_file = os.path.splitext(file_name)[0] + ".pdf"
            convert_word_to_pdf(file_name, output_file)
            print(f"Conversion completed. Download your PDF: {output_file}")
            files.download(output_file)
        else:
            print("Please upload a valid .docx file.")

# Call the function to upload and convert
upload_and_convert()
```
